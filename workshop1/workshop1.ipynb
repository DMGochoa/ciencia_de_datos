Por otro lado tenemos que la grafica no se ve muy bien y por tal motivo se va hacer en escala logaritmica. Vamos a construir histogramas y boxplots en escala logarítmica para las siguientes variables numéricas de interés:

````python
# Función para crear histogramas y boxplots en escala logarítmica para una variable dada
def plot_log_histogram_and_boxplot(data, column):
    fig, ax = plt.subplots(2, 1, figsize=(10, 8))

    # Histograma en escala logarítmica
    sns.histplot(data[data[column] > 0][column].apply(np.log), kde=True, ax=ax[0], bins=50)
    ax[0].set_title(f'Histograma (Escala Log) de {column}')
    ax[0].set_xlabel(f'Log({column})')
    ax[0].set_ylabel('Frecuencia')

    # Boxplot en escala logarítmica
    sns.boxplot(x=data[data[column] > 0][column].apply(np.log), ax=ax[1])
    ax[1].set_title(f'Boxplot (Escala Log) de {column}')
    ax[1].set_xlabel(f'Log({column})')

    plt.tight_layout()
    plt.show()

# Graficar INGRESOS OPERACIONALES en escala logarítmica
import numpy as np
plot_log_histogram_and_boxplot(df_raw, 'INGRESOS OPERACIONALES')
````

## Punto 6

Investiguen qué es el índice de correlación entre dos variables. Posteriormente, calculen el índice de correlación para todas sus variables numéricas. ¿Coincide este  ́ındice con las conclusiones que obtuvieron en el punto anterior?

El índice de correlación entre dos variables es una medida que indica la fuerza y la dirección de una relación lineal entre ellas. Los valores del índice de correlación varían entre -1 y 1. Un valor cercano a 1 implica una fuerte correlación positiva, es decir, cuando una variable aumenta, la otra también lo hace en una magnitud proporcional. Un valor cercano a -1 implica una fuerte correlación negativa, lo que significa que cuando una variable aumenta, la otra disminuye. Un valor cercano a 0 indica una correlación débil o ninguna correlación entre las variables.

La **discretización** es el proceso de convertir una variable numérica en una variable categórica o discreta. Esto se puede hacer por diversas razones, incluyendo facilitar la interpretación, reducir el ruido o preparar la variable para algoritmos específicos.

Se procede a discretizar la variable "INGRESOS OPERACIONALES" de dos maneras:

1. **Discretización No Supervisada**: Se dividirá la variable basándose en intervalos de igual ancho.
2. **Discretización Supervisada**: Se utilizará un algoritmo que considera la relación entre "INGRESOS OPERACIONALES" y otra variable. En este caso, se utilizará "GANANCIA (PÉRDIDA)" como la variable que guiará la discretización.

Comencemos con la discretización no supervisada.

````python
df_raw['IngresosOperacionalesCategorias'] = pd.qcut(df_raw['INGRESOS OPERACIONALES'], q=4, labels=['Bajo', 'Medio', 'Alto', 'Muy Alto'])
df_raw['IngresosOperacionalesCategorias']
````

````python
# Discretización No Supervisada
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier

# a. Discretización por Anchura Fija
max_value = df_raw["INGRESOS OPERACIONALES"].max()
df_raw['fixed_width_bins'] = pd.cut(df_raw['INGRESOS OPERACIONALES'], bins=np.arange(0, max_value+1, 1e10))

# b. Discretización por Frecuencia
df_raw['quantile_bins'] = pd.qcut(df_raw['INGRESOS OPERACIONALES'], q=3, duplicates='drop')

# c. Discretización por Clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(df_raw[['INGRESOS OPERACIONALES']])
df_raw['kmeans_bins'] = kmeans.labels_

# Discretización Supervisada

# a. Discretización de Árbol de Decisión
bin_edges = [-np.inf, 0, 10, np.inf]  # Ajusta estos bordes según tus datos
labels = ["Baja", "Media", "Alta"]
df_raw['GANANCIA_CAT'] = pd.cut(df_raw['GANANCIA (PÉRDIDA)'], bins=bin_edges, labels=labels)

# Ahora puedes usar DecisionTreeClassifier con la variable objetivo discretizada
y = df_raw["GANANCIA_CAT"]
decision_tree = DecisionTreeClassifier(max_depth=3, random_state=0)
decision_tree.fit(df_raw[['INGRESOS OPERACIONALES']], y)
df_raw['tree_bins'] = decision_tree.apply(df_raw[['INGRESOS OPERACIONALES']])

# Ver los resultados
resultados = df_raw[['INGRESOS OPERACIONALES', 'fixed_width_bins', 'quantile_bins', 'kmeans_bins', 'tree_bins', 'GANANCIA_CAT']].head(10)
resultados
````

La codificación de variables categóricas es un paso crucial en la preparación de datos para modelos que requieren entradas numéricas. Existen varias técnicas para codificar variables categóricas. La elección de la técnica depende en gran medida del tipo de variable categórica y del modelo que se va a utilizar.

**Tipos de codificación**:

1. **Codificación One-Hot (Dummy)**: Esta técnica crea una nueva columna para cada valor único en la variable categórica. Si la variable original tiene \( N \) valores únicos, la codificación one-hot resultará en \( N \) nuevas columnas binarias.

2. **Codificación Ordinal**: Si los valores categóricos tienen un orden (por ejemplo, bajo, medio, alto), puedes codificarlos en números que reflejen ese orden.

3. **Codificación Binaria**: Convierte cada entero a binario.

4. **Codificación de frecuencia o de tasa**: Reemplaza las categorías por la frecuencia o la tasa de ocurrencia de cada categoría.

5. **Embeddings (para redes neuronales)**: Las redes neuronales pueden aprender una representación densa de las categorías.

Para el conjunto de datos actual, una variable categórica interesante para codificar podría ser "MACROSECTOR", ya que agrupa a las empresas en sectores amplios y puede ser útil para modelar.

Dado que "MACROSECTOR" es una variable categórica nominal (sin orden inherente entre las categorías), la codificación one-hot sería apropiada. Sin embargo, si tuviéramos una variable categórica ordinal, como "nivel de satisfacción" con valores "bajo", "medio", "alto", la codificación ordinal sería más apropiada.

Ahora, procederé a codificar la variable "MACROSECTOR" usando la codificación one-hot.




````python
# Codificación One-Hot para la columna 'MACROSECTOR'
df_encoded = pd.get_dummies(df_raw, columns=['MACROSECTOR'], prefix=['MACROSECTOR'])
# Mostrando las columnas resultantes de la codificación one-hot
df_encoded.filter(regex='MACROSECTOR_').head(10)
````

La variable categórica "MACROSECTOR" ha sido codificada utilizando la técnica one-hot. Las columnas resultantes, como se muestra arriba, representan los diferentes macrosectores en los que las empresas pueden clasificarse. Cada empresa tiene un valor de 1 en la columna del macrosector al que pertenece y 0 en las demás columnas.

**Justificación**:
- La codificación one-hot es apropiada para variables categóricas nominales donde no hay un orden inherente entre las categorías.
- No se introducen supuestos numéricos o jerárquicos sobre las categorías, lo que es esencial si planeamos usar estos datos en un modelo.
- El tipo de atributo y el número de etiquetas únicas son importantes al elegir una estrategia de codificación. Por ejemplo, si tuviéramos una variable categórica con miles de categorías únicas, la codificación one-hot podría no ser ideal debido a la alta dimensionalidad que introduciría.

En conclusión, es fundamental elegir la técnica de codificación adecuada según el tipo de variable categórica y el modelo que se utilizará posteriormente. La codificación one-hot es una opción robusta para variables nominales, pero siempre es importante considerar el contexto y las características específicas de los datos al tomar decisiones de preprocesamiento.

````python
from sklearn.metrics import silhouette_score
# Determinar el número óptimo de clusters usando el método del codo
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.title('Método del Codo')
plt.xlabel('Número de clusters')
plt.ylabel('Inertia')
plt.show()
````

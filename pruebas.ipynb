```python
suma = 0

col = 'SEXO'
val = [0, 0, 0]

datos = {'Masculino': [], 'Femenino': [], 'Indeterminado': [], 'AÃ±o': []}

for i in range(8, 10):
    # Read the .sav file
    df = pd.read_csv(f'./data/nofetal200{i}.csv',
                     sep=',',
                     encoding='latin-1')
    result = df.groupby([col])['ANO'].count().sort_values(ascending=False)
    val[0] = val[0] + result.values[0]
    val[1] = val[1] + result.values[1]
    val[2] = val[2] + result.values[2]
    
    datos['AÃ±o'].append(f'200{i}')
    datos['Masculino'].append(result.values[0])
    datos['Femenino'].append(result.values[1])
    datos['Indeterminado'].append(result.values[2])
    
    print(result[0:5])

for i in range(10, 22):
    # Read the .sav file
    df = pd.read_csv(f'./data/nofetal20{i}.csv',
                     sep=',',
                     encoding='latin-1')
    result = df.groupby([col])['ANO'].count().sort_values(ascending=False)
    val[0] = val[0] + result.values[0]
    val[1] = val[1] + result.values[1]
    val[2] = val[2] + result.values[2]
    
    datos['AÃ±o'].append(f'20{i}')
    datos['Masculino'].append(result.values[0])
    datos['Femenino'].append(result.values[1])
    datos['Indeterminado'].append(result.values[2])
    
    print(result)

print(val[0]/14, val[1]/14, val[2]/14)
```
